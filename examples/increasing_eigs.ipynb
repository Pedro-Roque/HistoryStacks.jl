```julia
# Import packages
using Revise
using HistoryStacks
using Random
Random.seed!(421)
```

```julia
# Set up a random history stack
M = 10      # Maximum size of stack
δ = 1.0    # Tolerance for adding new data - needs to be large for this example to account for random data
H = HistoryStack(M, δ)
```

```julia
# Add random data to stack and look at stack eigenvalues
N = 30
λ = []
for i in 1:N
    update!(H, rand(2, 2), rand(2))
    push!(λ, HistoryStacks.eigmin(H))
end
```

```julia
# Plot the minimum eigenvalues over each iteration
using Plots
default(grid=false,framestyle=:box,fontfamily="Computer Modern",label="",lw=2,palette=:tab10)

plot(λ, xlabel="Iteration", ylabel="Minimum eigenvalue",xlims=(1,N))
vline!([M], c=:black, ls=:dash)
```
